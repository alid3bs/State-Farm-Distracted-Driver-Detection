```python
dic={}

for i in range(10):
    dic[f'c{i}']= len(os.listdir(os.path.join("data\\train", f'c{i}')))
```

```python
keys = list(dic.keys())
# get values in the same order as keys, and parse percentage values
vals = [dic[k] for k in keys]
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Bar(
    x=keys,
    y=vals,
)])
fig.update_layout(title_text='Number of images by Class')
```

```python
#first 400 pic as validation
import random
import shutil

for i in range(10):
    
    source = os.path.join(train_dir, f'c{i}')
    dest = os.path.join(validation_dir,f'c{i}')
    
    files = os.listdir(source)
    no_of_files = 400
    
    for file_name in random.sample(files, no_of_files):
        shutil.move(os.path.join(source, file_name), dest)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

activity_map = {'c0': 'Safe driving', 
                'c1': 'Texting - right', 
                'c2': 'Talking on the phone - right', 
                'c3': 'Texting - left', 
                'c4': 'Talking on the phone - left', 
                'c5': 'Operating the radio', 
                'c6': 'Drinking', 
                'c7': 'Reaching behind', 
                'c8': 'Hair and makeup', 
                'c9': 'Talking to passenger'}


plt.figure(figsize = (12, 20))
image_count = 1
train_direct = train_dir
for directory in os.listdir(train_direct):
    if directory[0] != '.':
        for i, file in enumerate(os.listdir(train_direct +'\\'+ directory)):
            if i == 1:
                break
            else:
                fig = plt.subplot(5, 2, image_count)
                image_count += 1
                image = mpimg.imread(train_direct +'\\'+ directory + '/' + file)
                plt.imshow(image)
                plt.title(activity_map[directory])
```
